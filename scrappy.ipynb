```python
bus_rt = []
bus_links = []
#collecting routes and route links


#will reuse this code
def collect_routes_links():
    bus_routes = driver.find_elements(By.XPATH,"//a[@class='route']")
    route_links = driver.find_elements(By.XPATH,"//div[@class='route_details']")
   
    if len(bus_routes) != len(route_links):
        print("error:mismatch between no of routes and number of links ")
    for i in range(len(bus_routes)):
        root = bus_routes[i].text
        bus_rt.append(root)
        
        links = route_links[i].find_elements(By.TAG_NAME,"a")
        for i in links:
            href = i.get_attribute("href")
            bus_links.append(href)
     # to convert data into dict       
           
#for page navigation
def page_navigation():
    
    page_number = 2
    while True:
        try:
            path = f"//div[@class= 'DC_117_pageTabs ' and contains(text(),'{page_number}')]"
            # path for pages
            #navigating to the second page
            next_page = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH,path)))
            ActionChains(driver).move_to_element(next_page).perform()
            time.sleep(2)
            next_page.click()
            time.sleep(2)
            collect_routes_links()
            # function called and data collected
            page_number = page_number + 1
            # incremented to the next page
        except Exception:
            break


transports = st_trans
for state in transports:
    driver = webdriver.Chrome()
    driver.get('https://www.redbus.in/online-booking/rtc-directory')
    time.sleep(2)
    driver.execute_script('window.scrollBy(0,200)')
    path = f"{state}"
   # routes = driver.find_element(By.LINK_TEXT,path).click()
    time.sleep(1)
    element = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.LINK_TEXT,path)))
    ActionChains(driver).move_to_element(element).click().perform()
    
    collect_routes_links()
    driver.execute_script('window.scrollBy(0,2000)')
    time.sleep(1)

    page_navigation() 
```

```python
def scroll_page():
    last_height = driver.execute_script('return document.body.scrollHeight')
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        #scroll down to the bottom
        time.sleep(1)
        new_height = driver.execute_script('return document.body.scrollHeight')
        #compare new_height with last_height
    
        if new_height == last_height :
            break
        last_height = new_height
```
